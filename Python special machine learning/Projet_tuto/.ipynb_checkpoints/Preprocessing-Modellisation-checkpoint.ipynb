### Encodage, Nettoyage
NB: l'encodage ne concerne que des variable qualitatives ou categorielles(les variables du type object dans pandas).
Alors en examinant nos variables du type object ci-dessous on se rend compte que ces variables contiennent quatre classes(ou categories) à savoir : **'negatieve'** et **'positive'** dans la variable **SARS-Cov-2 exam result**, **'not_detected'** et **'detected'** dans les autres variables du type object. L'encodage doit donc porter sur ces categories, en assignant à chaque categories d'une variable un code donné.

```python
for col in df.select_dtypes('object'):
    print(f'{col :-<50} {df[col].unique()}')
    
print('******************\n',df.dtypes.value_counts())
```

```python
def encodage(df):
    code = {'positive': 1,
        'negative': 0,
        'detected': 1,
        'not_detected': 0}
    
    for col in df.select_dtypes('object'):
        df[col] = df[col].map(code)
        
    return df
# on encode les categorie
```

```python
def feature_engineering(df):
    df['est malade'] = df[viral_columns].sum(axis=1) >=1
    df = df.drop(viral_columns, axis=1)
    return df
    
    
    
# feature_enginieering consiste à créer des variables à partir des variables existantes.
# ici on ajoute une nouvelle variable 'est malade', qui indique si le patient est positif à d'autre maladies(virus) et 
# on abandonne les variable du type viral(viral_columns) parce qu'elles ne sont pas trop importantes pour la prise de 
# decision(voir en bas la partie 'model.feature_importances_') . Cette 
#nouvelle variable peut etre utile pour la prise de decision.
```

```python
def imputation(df):
    #df['is na'] = (df['Parainfluenza 3'].isna()) | (df['Leukocytes'].isna())
    #df = df.fillna(-999)
    df = df.dropna(axis=0)
    return df

#ici on supprime les valeurs manquantes(NaN)
```

```python
def preprocessing(df):
    
    df = encodage(df)
    df = feature_engineering(df)
    df = imputation(df)
    
    X = df.drop('SARS-Cov-2 exam result', axis=1)
    y = df['SARS-Cov-2 exam result']
    
    print(y.value_counts())
    
    return X, y

#application des deux fonction encodage et imputation et separation de target et les autres features(X_train, y_train) 
#dans notre dataset. Ce dataset(à fournir à la fonction preprocessing) doit biensure être notre trainset
#(obtenu à travers train_test_split un peu plus haut, qui sera utilisé pour l'entrainement) 
#ou testset(pour la phase d'évaluation)
```

```python
X_train, y_train = preprocessing(trainset)
```

```python
X_train.dtypes.value_counts()
```

```python
X_test, y_test = preprocessing(testset)
```

```python
X_test.dtypes.value_counts()
```

### Premier modèle : Evaluation - Diagnostique

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import make_pipeline
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.preprocessing import PolynomialFeatures
from sklearn.decomposition import PCA
from sklearn.feature_selection import VarianceThreshold

# f_classif est pour le test de Annova
```

```python
model = make_pipeline(PolynomialFeatures(2), 
                      VarianceThreshold(),  # Retire les caractéristiques constantes pour eviter la division par 0
                      SelectKBest(f_classif, k=12),
                      RandomForestClassifier(random_state=0))

# NB: on peut associer dans un pipeline le selector, l'imputer et l'estimator.
#PolynomialFeatures(2): polynomial features de degré 2: permettra d'élévé chaque feature au carré
```

### - Procedure d'évaluation
Ici on essaie de definir les differentes metrics sur les quels l'évaluation du modèle va se baser.

La metric **F1 (f1_score)** est par exemple une très bonne **metric** pour avoir un bon aperçu entre le rapport de la **precision** et de la **sensibilité (recall)**.
La **precision** et le **recall** sont des metrics qui nous permettent de mesurer les proportion d'erreurs de **type 1** et d'érreurs de **type 2** dans un dataset, c'est à dire les proportion de **faux positif (false positive)** et de **faux négatif (false negative)** que le modèle éffectue 

```python
from sklearn.metrics import f1_score, confusion_matrix, classification_report
from sklearn.model_selection import learning_curve
```

```python
def evaluation(model):
    model.fit(X_train, y_train)
    ypred = model.predict(X_test)
    
    print(confusion_matrix(y_test, ypred))
    print(classification_report(y_test, ypred))
    
    N, train_score, val_score = learning_curve(model, X_train, y_train, 
                                               cv=4, scoring='f1', train_sizes=np.linspace(0.1, 1, 10))
    plt.figure(figsize=(12,8))
    plt.plot(N, train_score.mean(axis=1), label='train score')
    plt.plot(N, val_score.mean(axis=1), label='validation score')
    plt.legend()
    
# N : Contient les tailles des sous-échantillons utilisés, qui correspondent à chaque fraction dans train_sizes.
```

```python
evaluation(model)
```

#### model.feature_importances_

model.feature_importances_ : Cette propriété est spécifique aux modèles d’arbres de décision et aux modèles dérivés
(comme les forêts aléatoires ou les gradient boosting). feature_importances_ fournit un score pour chaque caractéristique
(ou variable) dans le modèle, indiquant combien chaque caractéristique a contribué à la prise de décision du modèle. 
Les valeurs sont normalisées pour que la somme des importances soit égale à 1. Par exemple, une caractéristique avec 
une importance de 0,3 a contribué plus que celle avec une importance de 0,1 dans les décisions du modèle.

cela nous permet de faire une bonne feature selection pour ne selectinner que des variable qui sont vraiment importante
pour la prise de decision et entrainer ainsi notre model avec ces variable.

#### que remarque-t-on sur la figure ci-dessous:

On remarque sur la figure ci-dessous que les variables du type **viral (viral_columns)** ne contribuent majoritairement pas à la prise de decision. 

**Alors :** 

   - 1) Ici on pourrait definir par exemple un seuil en dessous du quel toutes les variables sont ignorer pour ne selectionner que celles qui jou un très grand rôle pour la prise de decision.
   - 2) Au lieu de faire un **dropna(axis=0) sur tout le dataset** on pourrait tout simplement supprimer les colonnes du type **virus** (**viral_columns**) puis appliquer le **dropna()** uniquement sur les colonnes du type **blood** (**blood_columns**) cela nous permet d'avoir plus de données (**vue que le dropna() ne concerne dans ce cas que les colonnes du type blood**).

## Procedure d'évaluation

```python
from sklearn.metrics import confusion_matrix, f1_score, classification_report
from sklearn.model_selection import learning_curve
```

```python
def evaluation1(model):
    model.fit(X_train, y_train)
    ypred = model.predict(X_test)
    
    print(confusion_matrix(y_test, ypred))
    print(classification_report(y_test, ypred))
    
    N, train_score, val_score = learning_curve(model, X_train, y_train, 
                                               cv=4, scoring='f1', train_sizes=np.linspace(0.1, 1, 10))
    plt.figure(figsize=(12,8))
    plt.plot(N, train_score.mean(axis=1), label='train score')
    plt.plot(N, val_score.mean(axis=1), label='validation score')
    plt.title(model.steps[-1][-1]) # rends le nom du model comme title 
    plt.legend()
    plt.show()
```

### Precision Recall Curve

La fonction **precision_recall_curve** et la méthode **decision_function** sont deux outils de Scikit-Learn souvent utilisés dans le contexte de l’évaluation des modèles de classification, en particulier lorsque l’on travaille avec des **classes déséquilibrées**.

##### 1. precision_recall_curve
precision_recall_curve calcule la courbe de précision-rappel pour les modèles de classification binaire. Elle est souvent préférée à la courbe ROC lorsque les classes sont déséquilibrées, car elle met l’accent sur la capacité du modèle à identifier la classe positive.

Principe : Elle calcule les valeurs de précision et de rappel à différents **seuils de probabilité**, permettant de visualiser comment les deux métriques varient en fonction de ce seuil.
Utilisation : Cette courbe aide à choisir un seuil optimal, selon l'équilibre entre la précision (proportion de prédictions positives correctes) et le rappel (proportion d'éléments positifs correctement détectés).

##### 2. decision_function
La méthode decision_function fournit les scores de décision d’un modèle pour chaque instance, qui peuvent ensuite être utilisés comme seuils dans la courbe de **précision-rappel** ou pour la courbe ROC. Ces scores indiquent la “**confiance**” du modèle dans sa prédiction, mais ne sont pas des probabilités.

Principe : Contrairement à predict_proba, qui donne des probabilités, decision_function donne des scores bruts (positifs ou négatifs) pour chaque classe. Ces scores sont ensuite interprétés pour prédire la classe.
Utilisation : Elle est utile lorsque vous voulez un seuil de décision personnalisé ou analyser la confiance d'un modèle. Les modèles comme les SVM et les régressions logistiques utilisent souvent decision_function.

##### IMPORTANT : threshold

Définition : Un **seuil(threshold)** est **une valeur limite, au-delà de laquelle une instance est considérée comme appartenant à la classe positive**. Par exemple, si le seuil est **0.5**, toute instance avec un **score de décision supérieur à 0.5 sera classée comme positive**.

**Calcul de précision et de rappel** : En appliquant différents seuils, on peut voir comment changent les valeurs de précision et de rappel, ce qui permet de trouver un seuil optimal en fonction de l’objectif du modèle.
Utilisation : Lorsque l’on souhaite ajuster la performance du modèle pour favoriser soit la précision, soit le rappel, en fonction des besoins spécifiques (par exemple, réduire les faux positifs ou maximiser les vraies détections).

```python
from sklearn.metrics import precision_recall_curve
```

```python
precision, recall, threshold = precision_recall_curve(y_test, random.best_estimator_.decision_function(X_test))
```

```python
plt.plot(threshold, precision[:-1], label='Precision')
plt.plot(threshold, recall[:-1], label='Recall')
plt.legend()
```

```python
def model_final(model, X, threshold=0):
    return model.decision_function(X) > threshold
```

```python
y_pred = model_final(random.best_estimator_, X_test, threshold=-0.7)
```

```python
from sklearn.metrics import f1_score, recall_score, confusion_matrix
```

```python
f1_score(y_test, y_pred)
```

```python
recall_score(y_test, y_pred)
```
