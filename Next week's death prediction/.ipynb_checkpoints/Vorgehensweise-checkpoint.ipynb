```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Beispiel-Daten
data = {
    "PM2.5 (µg/m³)": [10, 20, 15, 30, 25, 40, 35, 50, 45, 60],
    "Asthma-Hospitalisierungen (pro 100.000)": [5, 10, 8, 20, 15, 30, 25, 40, 35, 50],
    "Bevölkerungsdichte (pro km²)": [100, 150, 120, 300, 250, 400, 350, 500, 450, 600],
    "Jahreszeit": ["Winter", "Sommer", "Frühling", "Winter", "Sommer", "Herbst", "Frühling", "Herbst", "Sommer", "Winter"]
}

# DataFrame erstellen
df = pd.DataFrame(data)

# Farbpalette für die Jahreszeiten
palette = {"Winter": "blue", "Sommer": "orange", "Frühling": "green", "Herbst": "red"}

# Scatterplot als Bubble Chart
plt.figure(figsize=(10, 7))
scatter = plt.scatter(
    x=df["PM2.5 (µg/m³)"],
    y=df["Asthma-Hospitalisierungen (pro 100.000)"],
    s=df["Bevölkerungsdichte (pro km²)"] / 2,  # Punktgröße
    c=df["Jahreszeit"].map(palette),          # Farbe
    alpha=0.7,
    edgecolor="k"
)

# Hinzufügen von Anmerkungen (optional)
for i, row in df.iterrows():
    plt.text(row["PM2.5 (µg/m³)"] + 0.5, row["Asthma-Hospitalisierungen (pro 100.000)"] + 0.5, 
             row["Jahreszeit"], fontsize=9, alpha=0.8)

# Titel und Achsen
plt.title("Zusammenhang zwischen PM2.5 und Asthma-Hospitalisierungen\n (Größe: Bevölkerungsdichte, Farbe: Jahreszeit)", fontsize=14)
plt.xlabel("PM2.5-Konzentration (µg/m³)", fontsize=12)
plt.ylabel("Asthma-Hospitalisierungen (pro 100.000)", fontsize=12)

# Legende für Farben erstellen
from matplotlib.lines import Line2D
legend_elements = [Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=10, label=season)
                   for season, color in palette.items()]
plt.legend(handles=legend_elements, title="Jahreszeit", loc="upper left", fontsize=10)

# Gitter und Layout
plt.grid(alpha=0.3)
plt.tight_layout()

# Plot anzeigen
plt.show()

```
