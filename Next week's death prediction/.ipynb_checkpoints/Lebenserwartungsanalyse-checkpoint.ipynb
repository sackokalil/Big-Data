Es bleibt immer noch offen zu bestimmen, welche Art von fehlenden Daten sich im Datensatz befinden.
Es gibt im Prinzip 3 Hauptarten von fehlenden Daten

1. Missing-completely-at-Random (MCAR): Wenn keine m철gliche Erkl채rung f체r das Fehlen der Daten eines Attributs gefunden 
werden kann.

2. Missing-at-Random (MAR): Wenn das Fehlen der Daten eines Attributs 체ber eine andere Spalte im Datensatz erkl채rt 
werden kann. (Ungl체ckliche Mitarbeiter teilen kein Feedback in einer Unternehmensfeedbackumfrage mit)

3. Missing-not-at-Random (MNAR): Wenn das Fehlen der Daten eines Attributs 체ber ein anderes, nicht im Datensatz
vorhandenes Attribut mithilfe dom채nes Wissens erkl채rt werden kann.

```python
# Anpassung linearer Regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import mean_squared_error, r2_score
from math import sqrt
```

```python
x_l철schen = df_l철schen.drop('Life expectancy',axis=1)
```

```python
y_l철schen = df_l철schen['Life expectancy']
y_l철schen = pd.DataFrame(y_l철schen)
```

```python
x_l철schen_train, x_l철schen_test, y_l철schen_train, y_l철schen_test =  train_test_split(x_l철schen,y_l철schen,test_size=.30)
```

```python
print("X_train: ",x_l철schen_train.shape)
print("X_test: ",x_l철schen_test.shape)
print("y_train: ",y_l철schen_train.shape)
print("y_test: ",y_l철schen_test.shape)
```

```python
linreg = LinearRegression()
linreg.fit(x_l철schen_train,y_l철schen_train)
```

```python
y_l철schen_pred = linreg.predict(x_l철schen_test)
```

```python
plt.scatter(y_l철schen_test,y_l철schen_pred)
plt.title('Vorhersage nach L철schen:')
plt.xlabel('Lebenserwartung_test')
plt.ylabel('Lebenserwartung_pred')
plt.show()
```

```python
# Evaluationsmetriken von linearer Regression
mse_l철schen = mean_squared_error(y_l철schen_test,y_l철schen_pred) # MSE-Wert
rmse_l철schen = sqrt(mse_l철schen) # RMSE-Wert
r2_l철schen = r2_score(y_l철schen_test, y_l철schen_pred) # R2-Score
n_l철schen = x_l철schen_test.shape[0]  # number of samples
k_l철schen = x_l철schen_test.shape[1]  # number of features
r2_adj_l철schen = 1 - ((1 - r2_l철schen) * (n_l철schen - 1) / (n_l철schen - k_l철schen - 1)) # Angepasster R2-score

LinReg_l철schen = pd.DataFrame({'Evalutionsmetriken': ['Mean-squared-error (MSE)','Root-mean-sqaured-error (RMSE)','R2-score','Angepasster R2-Score'],
                               'Scores': [mse_l철schen, rmse_l철schen, r2_l철schen, r2_adj_l철schen]})
print('Evaluationsergebnisse von linearer Regression nach L철schen fehlender Daten:\n')
LinReg_l철schen
```

### 2. Standard Scaling (Standardskalierung)
### Pr체fung auf Gaussche Verteilung in Datenspalten

```python
for column in df_mice_imputed.columns: 
    # Histogram
    sns.histplot(df_unscaled[column], kde=True)
    plt.title('Histogram der Spalte '+ column)
    plt.show()
```

 Aus Literaturrecherche ergibt sich die Tatsache, dass Standard-Skalierung zu einem Datensatz gut passt,
 wenn die Mehrheit der Attributen des Datensatzes eine Gausche Verteilung ergeben.
 Von den oberen Ergebnissen k철nnte festgestellt werden, dass die Mehrheit der Spalten im Datensatz
 keine Gausche Verteilung ergeben. 
 Daher wird Standard-skalierung nicht als ein passender Skalierungsansatz betrachtet.

### 3. Robust Scaling (Robust-Skalierung) vs. Power Transformation
### Fortgeschrittene Skalierverfahren

RobustScaler:
Diese Skalierung basiert seine Berechnung auf den Quartilwerten (Perzentile) der Daten, was robust gegen체ber Ausreiern ist.

Power-Transformation:
Die PowerTransformer-Klasse sch채tzt automatisch die beste Potenztransformation f체r die zu skalierenden Spalten ein.

```python
# Cross-Validation (Kreuz-Validierung) 체ber das Zielattribut zum Bestimmen des passendsten Skalierungsverfahrens
```

```python
from sklearn.model_selection import cross_val_score, KFold
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import RobustScaler, PowerTransformer
```

```python
X_scale = df_unscaled.drop('Life expectancy', axis=1)
y_scale = df_unscaled['Life expectancy']
```

```python
# Choose the models and scaling techniques
linRegressor = LinearRegression()
robustScaler = RobustScaler()
powerTransformer = PowerTransformer(method='yeo-johnson', standardize=True)
kfold = KFold(n_splits=5, shuffle=True, random_state=42)

X_scaled_RS = robustScaler.fit_transform(X_scale)
X_scaled_PT = powerTransformer.fit_transform(X_scale)

cv_RS = cross_val_score(linRegressor, X_scaled_RS, y_scale, cv=kfold, scoring='neg_mean_squared_error')
cv_PT = cross_val_score(linRegressor, X_scaled_PT, y_scale, cv=kfold, scoring='neg_mean_squared_error')

rmse_RS = np.sqrt(-cv_RS.mean())
rmse_PT = np.sqrt(-cv_PT.mean())

print('RMSE-Ergebnis vom Robust-Skalierer',rmse_RS)
print('\n')
print('RMSE-Ergebnis vom Power-Transformer',rmse_PT)
```

Robust-Skalierer ergibt einen niedrigeren RMSE-Ergebnis und wird daher f체r die Skalierung des Datensatzes genutzt.
