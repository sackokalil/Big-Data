Es bleibt immer noch offen zu bestimmen, welche Art von fehlenden Daten sich im Datensatz befinden.
Es gibt im Prinzip 3 Hauptarten von fehlenden Daten

1. Missing-completely-at-Random (MCAR): Wenn keine mögliche Erklärung für das Fehlen der Daten eines Attributs gefunden 
werden kann.

2. Missing-at-Random (MAR): Wenn das Fehlen der Daten eines Attributs über eine andere Spalte im Datensatz erklärt 
werden kann. (Unglückliche Mitarbeiter teilen kein Feedback in einer Unternehmensfeedbackumfrage mit)

3. Missing-not-at-Random (MNAR): Wenn das Fehlen der Daten eines Attributs über ein anderes, nicht im Datensatz
vorhandenes Attribut mithilfe domänes Wissens erklärt werden kann.

```python
linreg = LinearRegression()
linreg.fit(x_löschen_train,y_löschen_train)
```

```python
plt.scatter(y_löschen_test,y_löschen_pred)
plt.title('Vorhersage nach Löschen:')
plt.xlabel('Lebenserwartung_test')
plt.ylabel('Lebenserwartung_pred')
plt.show()
```

```python
# Evaluationsmetriken von linearer Regression
mse_löschen = mean_squared_error(y_löschen_test,y_löschen_pred) # MSE-Wert
rmse_löschen = sqrt(mse_löschen) # RMSE-Wert
r2_löschen = r2_score(y_löschen_test, y_löschen_pred) # R2-Score
n_löschen = x_löschen_test.shape[0]  # number of samples
k_löschen = x_löschen_test.shape[1]  # number of features
r2_adj_löschen = 1 - ((1 - r2_löschen) * (n_löschen - 1) / (n_löschen - k_löschen - 1)) # Angepasster R2-score

LinReg_löschen = pd.DataFrame({'Evalutionsmetriken': ['Mean-squared-error (MSE)','Root-mean-sqaured-error (RMSE)','R2-score','Angepasster R2-Score'],
                               'Scores': [mse_löschen, rmse_löschen, r2_löschen, r2_adj_löschen]})
print('Evaluationsergebnisse von linearer Regression nach Löschen fehlender Daten:\n')
LinReg_löschen
```

## Skalieren des Datensatzes

### 1. Min-Max Scaling (Minimum-Maximum Skalierung)

Da der multivariate Datensatz mehrere Attributen mit unterschiedlichen Wertebereichen enthält und Spalten wie Bevölkerung
'Adult Mortality' und 'Alkohol' starke Outliers haben, wurde Min-Max Scalierung nicht als das passendste Verfahren betrachtet.

### 2. Standard Scaling (Standardskalierung)
### Prüfung auf Gaußssche Verteilung in Datenspalten

```python
for column in df_mice_imputed.columns: 
    # Histogram
    sns.histplot(df_unscaled[column], kde=True)
    plt.title('Histogram der Spalte '+ column)
    plt.show()
```

 Aus Literaturrecherche ergibt sich die Tatsache, dass Standard-Skalierung zu einem Datensatz gut passt,
 wenn die Mehrheit der Attributen des Datensatzes eine Gaußsche Verteilung ergeben.
 Von den oberen Ergebnissen könnte festgestellt werden, dass die Mehrheit der Spalten im Datensatz
 keine Gaußsche Verteilung ergeben. 
 Daher wird Standard-skalierung nicht als ein passender Skalierungsansatz betrachtet.

### 3. Robust Scaling (Robust-Skalierung) vs. Power Transformation
### Fortgeschrittene Skalierverfahren

RobustScaler:
Diese Skalierung basiert seine Berechnung auf den Quartilwerten (Perzentile) der Daten, was robust gegenüber Ausreißern ist.

Power-Transformation:
Die PowerTransformer-Klasse schätzt automatisch die beste Potenztransformation für die zu skalierenden Spalten ein.

```python
from sklearn.model_selection import cross_val_score, KFold
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import RobustScaler, PowerTransformer
```

```python
X_scale = df_unscaled.drop('Life expectancy', axis=1)
y_scale = df_unscaled['Life expectancy']
```

```python
# Choose the models and scaling techniques
linRegressor = LinearRegression()
robustScaler = RobustScaler()
powerTransformer = PowerTransformer(method='yeo-johnson', standardize=True)
kfold = KFold(n_splits=5, shuffle=True, random_state=42)

X_scaled_RS = robustScaler.fit_transform(X_scale)
X_scaled_PT = powerTransformer.fit_transform(X_scale)

cv_RS = cross_val_score(linRegressor, X_scaled_RS, y_scale, cv=kfold, scoring='neg_mean_squared_error')
cv_PT = cross_val_score(linRegressor, X_scaled_PT, y_scale, cv=kfold, scoring='neg_mean_squared_error')

rmse_RS = np.sqrt(-cv_RS.mean())
rmse_PT = np.sqrt(-cv_PT.mean())

print('RMSE-Ergebnis vom Robust-Skalierer',rmse_RS)
print('\n')
print('RMSE-Ergebnis vom Power-Transformer',rmse_PT)
```

Robust-Skalierer ergibt einen niedrigeren RMSE-Ergebnis und wird daher für die Skalierung des Datensatzes genutzt.
